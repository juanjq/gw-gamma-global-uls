```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
```

```python
energies_m = [
    (0.15, 0.6), (0.6, 20), (0.15, 0.4), (0.3, 0.8), (0.5, 1.0), (1.0, 20), (0.15, 20)
]
r_95_m = [131.9, 63.0]
r_m = [38.1, -11.7]
s_m = [-0.99, 1.41]
ul_m = np.array([4.30e-11, 2.85e-12]) * u.Unit("1 / (s cm2)")
```

```python
energies_s = [
    (0.6, 20), (0.4, 0.8), (0.6, 1.0), (1.0, 20), (0.4, 20)
]
r_95_s = [-27.9, 2.9]
r_s = [-65.9, -65.4]
s_s = [-0.47, 1.29]
ul_m = np.array([3.28e-12, 1.30e-11]) * u.Unit("1 / (s cm2)")
```

```python
cm, cs = "b","r"

fig, ax = plt.subplots(figsize=(7, 4))

for i, (em, eM) in enumerate(energies_m[:2]):
    eC = np.sqrt(em * eM)
    ax.errorbar(eC, r_95_m[i], xerr=[[eC-em], [eM-eC]], color=cm, marker=".", capsize=3)
    err = ax.errorbar(eC, r_m[i], xerr=[[eC-em], [eM-eC]], color=cm, marker=".", capsize=3)
    err[-1][0].set_linestyle("--")

for i, (em, eM) in enumerate(energies_s[:2]):
    eC = np.sqrt(em * eM)
    ax.errorbar(eC, r_95_s[i], xerr=[[eC-em], [eM-eC]], color=cs, marker=".", capsize=3)
    err = ax.errorbar(eC, r_s[i], xerr=[[eC-em], [eM-eC]], color=cs, marker=".", capsize=3)
    err[-1][0].set_linestyle("--")

ax.plot([], [], color=cm, ls="-", marker=".", label="LST-only")
ax.plot([], [], color=cs, ls="-", marker=".", label="LST+MAGIC")
ax.plot([], [], color="gray", ls="-", marker="", label="95% region")
ax.plot([], [], color="gray", ls="--", marker="", label="Full Sky-Map")

ax.legend(loc=(1.03, 0), frameon=False)
ax.axhline(0, ls=":", color="lightgray")
ax.set_xlabel("Energy [TeV]"); ax.set_ylabel("Fraction of Sky-Map ULs [%]")
ax.set_xscale("log")
plt.show()
```
